#Subsitting Data : Columns
Sometimes while dealing with large number data we have so many columns that we actually dont need
we can provide list of columns to the dataFrame of our data and the data will have only that cols

```python
# we can list down the columns of our data by the method columns
# mydf.columns
# we can override our dataframe with the cols we want to include
mydf = mydf[['title', 'description', 'publishTime', 'duration_seconds',
       'viewCount', 'likeCount', 'commentCount']]
```

```python
mydf.head()
#The dataframe will only include these cols
```

#Subsitting the Rows
we can specify a condition for a rows of a specific column
for example we can specify to include only those rows that has value of higher then 1Million view count

```python
# Subsiting using loc
mydf_subsit1 =mydf.loc[mydf['viewCount'] >1_000_000]
mydf_subsit1
```

#Subsitting using Query
this is a new method introduce to Pandas it is cleaner to write then the loc method

```python
mydf_subsit2 = mydf.query('viewCount > 1000000')
mydf_subsit2
```

#Data Cleansing
#Looking for missing Values
an important factor for dealing with the data is to understand the missing values in the data or Null and dropping it
if you run the na on loc method with ~ it will only store the null values and with this it will do the opposite and remove the null values

```python

mydf['viewCount'].isna() #this will tell us True for the row the is null
mydf = mydf.loc[~mydf['viewCount'].isna()]
mydf = mydf.loc[~mydf['likeCount'].isna()]
```

```python
mydf['viewCount'].isna()
# upon checking we dont have any null values
```

#Casting dtype
what it means is that changing the type of col to some different the it already is
As a Rule any numeric col that has null values in it is of type float since we have dropped the null values we can cast it to integer
we can do this by calling the method as type on that specific col and specifying the type to which we wanna cast
happy

```python
# running dtypes on dataframe will  tell us what type of cols we have
# mydf.dtypes
mydf['viewCount'] = mydf['viewCount'].astype('int')
mydf['likeCount'] = mydf['likeCount'].astype('int')
```

```python
# Now there is an interesting col publish time which is of type obj lets convert it to
#  datetime
mydf['publishTime'] = pd.to_datetime(mydf['publishTime'])

```

```python
mydf.dtypes
```
